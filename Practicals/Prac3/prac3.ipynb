```python
import pandas as pd
import numpy as np
import seaborn as sb
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
N_OF_DATA = 20

xs_random = np.random.rand(N_OF_DATA)
xs_training = xs_random * 5
ys = 2 * np.sin(1.5 * xs_training)

# add noise
mu, sigma = 0, 1
ys_noise = ys + np.random.normal(mu, sigma, N_OF_DATA)

xs_random_test = np.random.rand(N_OF_DATA)
xs_test = xs_random_test * 5
ys_test = 2 * np.sin(1.5 * xs_test) + np.random.normal(mu, sigma, N_OF_DATA)

reshaped_xs_test = xs_test.reshape(-1, 1)
```

```python
sb.scatterplot(x=xs_training, y=ys_noise, color='red')
sb.lineplot(x=xs_training, y=ys)
```

### Q1

```python
poly_reg_model = LinearRegression()


poly2_reg_model.fit(xs_training.reshape(-1,1), ys_noise)
y_predicted = poly2_reg_model.predict(reshaped_xs_test)
sse = np.sum((ys_noise - y_predicted) ** 2)
sb.scatterplot(x=xs_training, y=ys_noise, color='red')
sb.lineplot(x=xs_test, y=y_predicted)
```

```python
poly2_reg_model = LinearRegression()
poly2 = PolynomialFeatures(degree=2, include_bias=False)

poly2_features = poly2.fit_transform(xs_training.reshape(-1, 1))
poly2_reg_model.fit(poly2_features, ys_noise)

y2_predicted = poly2_reg_model.predict(poly2.fit_transform(reshaped_xs_test))
sse2 = np.sum((ys_noise - y2_predicted) ** 2)

sb.scatterplot(x=xs_training, y=ys_noise, color='red')
sb.lineplot(x=xs_test, y=y2_predicted)
```
