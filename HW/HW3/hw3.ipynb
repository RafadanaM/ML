### No 1

```python
x = np.array([0.0000, 0.5236, 1.0472, 1.5708, 2.0944, 2.6180, 3.1416])
t = np.array([2.7600, -2.3902, 1.3800, 0.0000, -1.3800, 2.3902, -2.7600])
mean_t = t.mean()
```

```python
# Scatter
sb.scatterplot(x=x, y=t)
```

```python
# Polynomial a

poly = PolynomialFeatures(degree=3, include_bias=False)
reshaped_x = x.reshape(-1, 1)
poly_features = poly.fit_transform(reshaped_x)
poly_reg_model = LinearRegression()

poly_reg_model.fit(poly_features, t)

y_predicted = poly_reg_model.predict(poly_features)

ssr_a = np.sum((y_predicted - t)**2)
sst_a = np.sum((t - mean_t)**2)

r_squared_a = 1 - (ssr_a / sst_a)
print('R squared is : ' + str(r_squared_a))

sb.scatterplot(x=x, y=t, color='red') 
sb.lineplot(x=x, y=y_predicted)
```

```python
# Polynomial b

poly = PolynomialFeatures(degree=10, include_bias=False)
reshaped_x = x.reshape(-1, 1)
poly_features = poly.fit_transform(reshaped_x)
poly_reg_model = LinearRegression()

poly_reg_model.fit(poly_features, t)

y_predicted = poly_reg_model.predict(poly_features)

ssr_b = np.sum((y_predicted - t)**2)
sst_b = np.sum((t - mean_t)**2)

r_squared_b = 1 - (ssr_b / sst_b)

print('R squared is : ' + str(r_squared_b))

sb.scatterplot(x=x, y=t, color='red') 
sb.lineplot(x=x, y=y_predicted)
```

```python
iris = pd.read_csv("./iris.csv", header=None)
iris_copy = iris.loc[:, [0, 4]]
iris_copy.columns = ['a', 'b']
def posterior(training_data: pd.DataFrame, k, x, p):
    # print(np.unique(training_data[1]))
    # if len(np.unique(training_data[-1]) != k):
    #     raise Exception('unique class in training_data is not equal to k')
    ## separate data to each classifier

    # mean = training_data.groupby(by=['b']).mean()
    # std = training_data.groupby(by=['b']).std()
    grouped = training_data.groupby('b')['a'].describe().reset_index()
    
    # pxa = []
    ll_dict = {}
    px = 0
    for i in range(k):
        current_class = grouped.iloc[[i]]['b'].values[0]
        mean_x  = grouped.iloc[[i]]['mean'].values[0]
        std_x = grouped.iloc[[i]]['std'].values[0]
        print(current_class)
        ll = (1/(np.sqrt(2*np.pi)*std_x)) * np.exp(-(x-mean_x)**2/(2*std_x**2))
        ll_dict[current_class] = ll
        px += ll * p
    

        # pxa += px
        # product = np.prod(ll)
        # postr = product * (1 / k)
        # print(postr)

        #sum dari ll * Pc
    result_dict = {}

    for key in ll_dict:

        result_dict[key] = (ll_dict[key] * p) / px
    
        sb.lineplot(x, result_dict[key])
    return result_dict

  

x = np.linspace(-5,10,1501)
result = posterior(iris_copy, 3, x, 1/3)

print(result)



# sb.lineplot(x=x, y=result['Iris-setosa'])
```
